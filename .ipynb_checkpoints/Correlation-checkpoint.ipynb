```python
def standard_units(x):
    '''Convert any array of numbers to standard units'''
    return (x - np.average(x))/np.std(x)
```

```python
Table().with_columns('mpg (SU)', standard_units(suv.column('mpg')), 'msrp (SU)', standard_units(suv.column('msrp'))).scatter(0, 1)
plots.xlim(-3, 3)
plots.ylim(-3, 3);
```

# Correlative coefficient

```python
x = np.arange(1, 7, 1)
y = make_array(2, 3, 1, 5, 2, 7)
t = Table().with_columns('x', x, 'y', y)
t
```

```python
t.scatter('x', 'y', s = 30, color = 'red')
```

```python
t = t.with_columns('x (SUnits)', standard_units(x), 'y(Sunits)', standard_units(y))
t
```

```python
su_products = t.column(2) * t.column(3)
t = t.with_column('product of standard units', su_products)
t
```

```python
r = np.average(t.column(4))
r
```

```python
def correlation_coefficient_r(tb1, x, y):
    '''tb1 is a tabe, x and y are column labels'''
    x_in_standardunits = standard_units(tb1.column(x))
    y_in_standardunits = standard_units(tb1.column(y))
    return np.average(x_in_standardunits * y_in_standardunits)
```

```python
correlation_coefficient_r(t, 'x', 'y') #to check if the function works :)
```

```python
correlation_coefficient_r(suv, 'mpg', 'msrp')
```
